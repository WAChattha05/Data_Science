```python
import pandas as pd
import requests
import numpy as np

final = pd.DataFrame()  # Initialize the final DataFrame

for j in range(1, 501):
    response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=e17334c802321c96ce889d67d24dd07c&language=en.US&page={}'.format(j)).json()

    popularity = []
    vote_count = []
    original_language = []
    title = []
    overview = []
    release_date = []
    vote_average = []

    for i in response['results']:
        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        original_language.append(i['original_language'])
        title.append(i['title'])
        overview.append(i['overview'])
        release_date.append(i['release_date'])
        vote_average.append(i.get('vote_average', np.nan))  # Assign NaN for missing values

    d = {'title': title, 'overview': overview, 'original_language': original_language, 'release_date': release_date, 'popularity': popularity, 'vote_count': vote_count, 'vote_average': vote_average}
    df = pd.DataFrame(d)

    final = pd.concat([final, df], ignore_index=True)
```
